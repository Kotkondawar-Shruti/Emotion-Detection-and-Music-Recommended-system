```python
import pandas as pd

# 1. Load the file
df = pd.read_csv("Music_Info.csv")

# 2. Define words you consider moods
mood_keywords = [
    'happy','sad','energetic','calm','chill','romantic','angry','melancholic',
    'uplifting','dark','moody','peaceful','relaxed','excited','dreamy','party'
]

# 3. Function to extract all mood words from the tags column
def extract_moods(tag_string):
    if pd.isna(tag_string):
        return None
    tags = [t.strip().lower() for t in tag_string.split(',')]
    moods_found = [t for t in tags if any(m in t for m in mood_keywords)]
    return ', '.join(moods_found) if moods_found else None

# 4. Create a new column 'mood'
df['mood'] = df['tags'].apply(extract_moods)

# 5. Save a new CSV (all original columns + new mood column)
df.to_csv("Music_Info_with_Mood.csv", index=False)

print("Done! New file saved as Music_Info_with_Mood.csv")
print(df.head())
```
